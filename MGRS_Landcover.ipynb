```python
from go_utils import get_api_data

lc_df = get_api_data("land_covers")
lc_df
```

```python
lc_df = lc_df[(lc_df["lc_Latitude"] < 90) & (lc_df["lc_Latitude"] > -90) & (lc_df["lc_Latitude"] < 180) & (lc_df["lc_Latitude"] > -180)]
lc_df
```

# Drop expanded columns

```python
lc_df = lc_df.drop(columns=[col for col in lc_df.columns if col.count("_") == 2])
```

```python
lc_df
```

```python
import mgrs

def get_mgrs(lat, lon):
    m = mgrs.MGRS()
    c = m.toMGRS(lat, lon)
    return m.toLatLon(c)
```

```python
import numpy as np
```

```python
vectorized_mgrs = np.vectorize(get_mgrs)
```

```python
latitude, longitude = vectorized_mgrs(lc_df["lc_Latitude"].to_numpy(), lc_df["lc_Longitude"].to_numpy())
```

```python
lc_df["lc_MGRSLatitude"] = latitude
```

```python
lc_df["lc_MGRSLongitude"] = longitude
```

# Build buffered DF

```python
import geopandas as gpd
glc_df = gpd.GeoDataFrame(lc_df, geometry=gpd.points_from_xy(lc_df["lc_MGRSLongitude"], lc_df["lc_MGRSLatitude"]), crs="EPSG:4326").to_crs(epsg=3395)
```

```python
glc_df
```

```python
glc_df.geometry = glc_df.buffer(50, cap_style=3)
```

```python
glc_df = glc_df.to_crs(epsg=4326)
```

```python
glc_df = glc_df.drop(["lc_GlobeTeams"], axis=1)
```

```python
glc_df
```

# Removing unprintable characters

```python
from string import printable
import pandas as pd

for index, row in glc_df.iterrows():
    for col in glc_df.columns:
        item = row[col]
        if not pd.isna(item):
            for bad in set(str(item)).difference(printable):
                #print(bad)
                item = item.replace(bad, "")
                #print(item)

            glc_df.at[index, col] = item
            if set(str(item)).difference(printable):
                print(glc_df.loc[index, col])
                #print(item)
```

# Visualization

```python
import folium
map_l8 = folium.Map(zoom_start=15)
for _, r in glc_df[:1000].iterrows():
    #without simplifying the representation of each borough, the map might not be displayed
    #sim_geo = gpd.GeoSeries(r['geometry'])
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'black'})
    geo_j.add_to(map_l8)
```

```python
import ee

ee.Initialize()
```

```python
dataset = ee.ImageCollection("ESA/WorldCover/v100").first()
```

```python
visualization = {
  "bands": ['Map'],
}
```

```python
def add_ee_layer(self, ee_object, vis_params, name):
    
    try:    
        # display ee.Image()
        if isinstance(ee_object, ee.image.Image):    
            map_id_dict = ee.Image(ee_object).getMapId(vis_params)
            folium.raster_layers.TileLayer(
            tiles = map_id_dict['tile_fetcher'].url_format,
            attr = 'Google Earth Engine',
            name = name,
            overlay = True,
            control = True
            ).add_to(self)
        # display ee.ImageCollection()
        elif isinstance(ee_object, ee.imagecollection.ImageCollection):    
            ee_object_new = ee_object.mosaic()
            map_id_dict = ee.Image(ee_object_new).getMapId(vis_params)
            folium.raster_layers.TileLayer(
            tiles = map_id_dict['tile_fetcher'].url_format,
            attr = 'Google Earth Engine',
            name = name,
            overlay = True,
            control = True
            ).add_to(self)
        # display ee.Geometry()
        elif isinstance(ee_object, ee.geometry.Geometry):    
            folium.GeoJson(
            data = ee_object.getInfo(),
            name = name,
            overlay = True,
            control = True
        ).add_to(self)
        # display ee.FeatureCollection()
        elif isinstance(ee_object, ee.featurecollection.FeatureCollection):  
            ee_object_new = ee.Image().paint(ee_object, 0, 2)
            map_id_dict = ee.Image(ee_object_new).getMapId(vis_params)
            folium.raster_layers.TileLayer(
            tiles = map_id_dict['tile_fetcher'].url_format,
            attr = 'Google Earth Engine',
            name = name,
            overlay = True,
            control = True
        ).add_to(self)
    
    except Exception as e:
        print("Could not display {}".format(name))
        print(str(e))
        
folium.Map.add_ee_layer = add_ee_layer
```

```python
map_l8.add_ee_layer(dataset, visualization, "Worldcover Map")
folium.LayerControl().add_to(map_l8)
```
